```python
# Function to format column names
def format_column_names(df):
    df.columns = df.columns.str.strip()  # Remove leading/trailing spaces
    df.columns = df.columns.str.replace('_', ' ')  # Replace underscores with spaces
    df.columns = df.columns.str.title()  # Capitalize first letter of each word
    return df

# Apply the formatting function
df = format_column_names(df)

# Display the DataFrame with renamed columns
print(df)
```

```python
df['year'] = pd.to_datetime(df['Created Date'])

# Convert year to integer type, keeping NaNs
df['year'] = df['year'].dt.year.astype('Int64')  # 'Int64' is pandas' nullable integer type

# excluding the rows where the 'CreatedYear' column is either 0 or 1970
df = df[ df['year'] != 1970]
df = df[df['year'] != 0]
df.head()
```

## Plotting the chart for Top10 Youtubers in terms of Subscribers

```python
# Sort by subscribers and select top 10
top_10 = df.sort_values('Subscribers', ascending=False).head(10)

# Create a color palette
colors = plt.cm.viridis(np.linspace(0, 1, 10))

# Plot
plt.figure(figsize=(10,7))
bars = plt.bar(top_10['Youtuber'], top_10['Subscribers'], color=colors)

# Add labels and title
plt.ylabel('Subscribers')
plt.xlabel('Youtubers')
plt.title('Top 10 YouTubers by Subscribers')

# Add a label on top of each bar indicating the number of subscribers
for bar in bars:
    yval = bar.get_height()
    yval_millions = yval / 1_000_000  # Convert to millions
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05*yval, f'{yval_millions:.2f}M', ha='center', va='bottom')

# Rotate x-axis labels if they overlap
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
